```python
from sklearn.externals import joblib
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model_dir = str(os.path.join("..","..","models"))
data_dir = str(os.path.join("..", "..", "data_dir", "twitter_naive_bayes"))
text_clf = joblib.load(str(os.path.join(model_dir, "naive_bayes.plk"))) 
```

```python
#getting data
tweets = []

with open(str(os.path.join("..", "..", "data_dir", "twitter_naive_bayes", "bmw_relevance_test_all.json")), 'r',
         encoding='latin-1') as f:
    for line in f:
        tweets.append(json.loads(line))

#data = pd.read_json(str(os.path.join("..", "..", "data_dir", "twitter_naive_bayes", "bmw_relevance_test.json")))
#tweets = list(data["processed_text"])
```

```python
X_test = []
for doc in tweets:
    X_test.append(" ".join(doc["processed_text"]))
```

```python
count_vectorizer = joblib.load(str(os.path.join(model_dir, "count_vectorizer.plk")))
tf_transformer = joblib.load(str(os.path.join(model_dir, "tf_transformer.plk")))

X_test_counts = count_vectorizer.transform(X_test)
X_test_tf = tf_transformer.transform(X_test_counts)
```

```python
predicted = text_clf.predict(X_test_tf)
```

```python
#for doc, category in zip(X_test, predicted):
#    print('%r => %s' % (doc, category))
```

```python
df = pd.DataFrame(data={"processed_text":X_test,"relevance":predicted})
df.to_csv(str(os.path.join(data_dir, "bmw_relevant_results_nb_ulysses.csv")))
```
