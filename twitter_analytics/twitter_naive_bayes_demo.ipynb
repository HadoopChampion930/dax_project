```python
from sklearn.externals import joblib
import pandas as pd
from pymongo import MongoClient
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
client = MongoClient("mongodb://igenie_readwrite:igenie@35.189.101.142:27017/dax_gcp")
db = client["dax_gcp"]
collection = db["tweets"]
```

```python
text_clf = joblib.load(r"/Users/Ulysses/Desktop/Desktop/dax/model/naive_bayes.plk") 
```

```python
#getting data
cursor = collection.find({"constituent":"adidas"}, {"text":1,"processed_text":1}).limit(10000)
tweets = list(cursor)
```

```python
X_test = []
for doc in tweets:
    X_test.append(" ".join(doc["processed_text"]))
```

```python
count_vectorizer = joblib.load(r"/Users/Ulysses/Desktop/Desktop/dax/model/" + "count_vectorizer.plk")
tf_transformer = joblib.load(r"/Users/Ulysses/Desktop/Desktop/dax/model/" + "tf_transformer.plk")

X_test_counts = count_vectorizer.transform(X_test)
X_test_tf = tf_transformer.transform(X_test_counts)
```

```python
predicted = text_clf.predict(X_test_tf)
```

```python
for doc, category in zip(X_test, predicted):
    print('%r => %s' % (doc, category))
```

```python
df = pd.DataFrame(data={"processed_text":X_test,"relevance":predicted})
```

```python
t = df[df["relevance"] == 1]
```
